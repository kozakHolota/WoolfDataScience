```python
import sympy as sp
from sympy.plotting import plot
from sympy import symbols
import matplotlib.pyplot as plt
```

```python
x = symbols('x')
f = x**2 + 2*x - 1
plot(f, (x, -5, 5), title='Графік f(x) = x² + 2x - 1')
```

```python
f1 = sp.sin(x)
f2 = sp.cos(x)

p = plot(f1, f2, (x, -5, 5),
         title='Графіки sin(x) та cos(x)',
         legend=True,
         show=False)

p[0].label = 'sin(x)'
p[1].label = 'cos(x)'

p.show()
```

```python
f = sp.exp(-x/2) * sp.sin(2*x)

p = plot(f, (x, -2*sp.pi, 2*sp.pi),
         title='Затухаючі коливання',
         xlabel='x',
         ylabel='f(x)',
         line_color='blue',
         show=False)

p[0].line_color = 'red'
p.show()
```

```python
from sympy.plotting import plot3d
from sympy import symbols

x, y = symbols('x y')
f = sp.sin(x) * sp.cos(y)

plot3d(f, (x, -5, 5), (y, -5, 5),
       title='3D поверхня f(x,y) = sin(x)cos(y)')
```
