# Завдання №1

### Позначимо:
- $ x_1 $ — кількість розроблених сайтів  
- $ x_2 $ — кількість розроблених інтернет-магазинів  
- $ x_3 $ — кількість інтеграцій з ERP  

---

### Цільова функція максимізації прибутку:

$$
f(x_1, x_2, x_3) = 2000x_1 + 9000x_2 + 6000x_3 \rightarrow \max
$$

---

### Система обмежень:

- Дизайнерів — 2, фонд праці кожного — 160 годин  
  Отже, сумарно: $ 2 \\cdot 160 = 320 $ годин  
  Обмеження:  
  $$
  12x_1 + 6x_2 + 2x_3 \leq 320
  $$

- Програмістів — 2, фонд праці одного — 96 годин  
  Отже, сумарно: $ 2 \\cdot 96 = 192 $ годин  
  Обмеження:  
  $$
  12x_1 + 24x_2 + 18x_3 \leq 192
  $$

- Тестувальник — 1, фонд праці — 180 годин  
  Обмеження:  
  $$
  12x_1 + 18x_2 + 12x_3 \leq 180
  $$

- Умови невід’ємності:
  $$
  x_1 \geq 0,\quad x_2 \geq 0,\quad x_3 \geq 0
  $$

```python
from scipy.optimize import linprog

# Цільова функція (мінімізація, тому беремо -f)
c = [-2000, -9000, -6000]  # мінімізуємо -f(x), бо linprog працює тільки на мінімум

# Матриця нерівностей (A_ub * x <= b_ub)
A = [
    [12, 6, 2],
    [12, 24, 18],
    [12, 18, 12]
]

"""
Дизайнерів 2, час кожного - 160, отже разом - 320
Програмістів 2 але за умовою їх загальний час - 192
Тестер один і при смерті від перепрацювання, а отже його час за умовою - 180
"""
b = [320, 192, 180]

# Межі змінних: x1, x2 ≥ 0
bounds = [(0, None), (0, None), (0, None)]

# Розв’язання
result = linprog(c, A_ub=A, b_ub=b, bounds=bounds, method='highs')

if result.success:
    x1, x2, x3 = result.x
    print(f"Оптимальне значення x1 (Сайти): {x1:.2f}")
    print(f"Оптимальне значення x2 (Інтернет-магазин): {x2:.2f}")
    print(f"Оптимальне значення x3 (Інтеграція з ERP): {x3:.2f}")
    print(f"Максимальний прибуток: {-result.fun:.2f}")
else:
    print("Помилка:", result.message)
```

Проаналізувавши розвʼязок, можна ствердити, що для компанії сенсу нема як у створенні сацтів, так і в інтеграції з ERP. Прибуток приносить лише створення Інтернет магазинів.
