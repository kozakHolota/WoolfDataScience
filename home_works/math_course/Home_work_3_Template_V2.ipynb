Дано 4 точки: $M_0(0, 0, 0), M_1(1, \frac{1}{3}, 0), M_2(0, 2, \frac{1}{4}), M_3(\frac{1}{2}, \frac{1}{2}, 1)$.

Також дано 3 вектори, які задають паралелепіпед із зазначеними точками:
- $\overline{a}=\overline{M_0M_1}$
- $\overline{b}=\overline{M_0M_2}$
- $\overline{c}=\overline{M_0M_3}$

Порахуй координати векторів через задані точки.  
Порахуй об'єм, площу повної поверхні паралелепіпеда та кути між ребрами паралелепіпеда, координати решти вершин паралелепіпеда.



$\overline{a} = \overline{M_0M_1} = \left (\begin{array}{cc}
1\\
\frac{1}{3}\\
0
\end{array}\right)
\\
\overline{b} = \overline{M_0M_2} = \left (\begin{array}{cc}
0\\
2\\
\frac{1}{4}
\end{array}\right)
\\
\overline{c} = \overline{M_0M_3} = \left (\begin{array}{cc}
\frac{1}{2}\\
\frac{1}{2}\\
1
\end{array}\right)
$

```python
import numpy as np
import matplotlib.pyplot as plt

# Створюємо вектори
vectors = [
    np.array([1, 1/3, 0]),
    np.array([0, 2, 0.25]),
    np.array([0.5, 0.5, 1])
]

# Створюємо графік
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Початкова точка для всіх векторів
origin = np.array([0, 0, 0])

# Кольори для векторів
colors = ['r', 'g', 'b']

# Малюємо вектори
for i, v in enumerate(vectors):
    ax.quiver(origin[0], origin[1], origin[2],  # початкова точка
              v[0], v[1], v[2],                 # напрямок
              color=colors[i],                  # колір з попередньо визначеного списку
              alpha=0.8,
              lw=2)

# Налаштування графіка
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Вектори')

# Встановлюємо однакові масштаби для всіх осей
ax.set_box_aspect([1,1,1])

plt.show()
```
